```python
# construct wordcloud function
def visualize_wordcloud_dict_frequencies(dict_freqs, title, relative_scaling=0.5, max_words=100,
                               background_color='black'):
   plt.figure(figsize=(10, 10))
   wordcloud = WordCloud(width=900, height=500, max_words=max_words, relative_scaling=relative_scaling,
                         normalize_plurals=False, background_color=background_color).generate_from_frequencies(
       dict_freqs)
   plt.title(title)
   plt.imshow(wordcloud, interpolation='bilinear')
   plt.axis("off")
   plt.show()
```

```python
df=pd.concat([data.Translated_Review,data.Sentiment],axis=1)
df.dropna(axis=0,inplace=True)
df.head(10)
```

```python
df.shape
```

```python
df2=df.rename(columns={'Translated_Review': 'Review'})
```

```python
start = time.time()

# TfidfVectorizer: Convert a collection of raw documents to a matrix of TF-IDF features.
#Equivalent to CountVectorizer followed by TfidfTransformer.
vectorizer = TfidfVectorizer(max_df=1000, min_df=100)

X_tfidf = vectorizer.fit_transform(df2['Review'])

end = time.time()
print("It took {} sec to fit and transform all documents.".format(end - start))
```

```python
# Calculate the cosine similarity:
#Cosine Similarity will generate a metric that says how related are two documents by looking 
#at the angle instead of magnitude, like in the examples below:
#Cosine similarity is not only telling the similarity between 
#two vectors but it also test for orthogonality of vector. 
#https://medium.com/@paritosh_30025/natural-language-processing-text-data-vectorization-af2520529cf7
#http://blog.christianperone.com/2013/09/machine-learning-cosine-similarity-for-vector-space-models-part-iii/
```

```python
# The tfidf_matrix[0:1] is the Scipy operation to get the first row of the sparse matrix 
# and the resulting array is the Cosine Similarity between the first document with all documents in the set
cosine_similarity(X_tfidf[0:1], X_tfidf)
```
