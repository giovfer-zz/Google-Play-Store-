# App ratings across categories - One Way Anova Test


```python
# Category and Type (paid or free)
```

```python
# % of paid versus free
# Visualize the contingency table between category and type
tablepaidcat= pd.crosstab(data['Category'], data['Type'],normalize='index')
print(tablepaidcat)
```

```python
# Heatmap Category and Type
_=plt.figure(figsize=(12,8))
_ =sns.heatmap(pd.crosstab(data['Category'], data['Type'], normalize='index'),annot=True, vmin=-1, cmap="YlGnBu",linewidths=.5) 
_=plt.title('Contigency table Apps categories by Type')
```

Chi square test: use the chi-square test for independence to determine whether there is a 
significant relationship between two categorical variables.

```python
# Chi square between category and Type
chi, pvalue, dof, expctd = chi2_contingency(pd.crosstab(data['Category'], data['Type']))
print("Chi-square between {} and {} is {} with a p-value of {:2f}".format('Category', 'Type', chi, pvalue))
```

```python
#The variable categorical is highly associated with the variable Type, the null hypothesis is rejected
```
